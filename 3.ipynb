```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(1,11)
y1 = np.power(X,2)
y2 = np.power(X,3)
plt.plot(X,y1, color='blue', label='X Squared')
plt.plot(X,y2,color='red', label='X Cubed')
plt.title("Basic line plot")
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# create a bar plot
categories = ['A', 'B', 'C', 'D']
values = [23,45,56,78]
plt.figure(figsize=(6,4))
plt.bar(categories,values,color='green')    
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title("basic bar plot")
plt.show()
```

```python
#plot a histogram 
data = np.random.randn(1000)
plt.hist(data,bins=30,color='pink',edgecolor='black')
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot a scatter plot
X = np.arange(1,101)
y = 0.1*X*X + 0.3
plt.scatter(X, y, color='red', alpha=0.5, s=8, marker='+')
plt.title("Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
sizes = [20,30,40,10]
labels = ['apple','banana','orange','grapes']
colors = ['red','yellow','orange','green']
plt.pie(sizes, labels= labels, colors = colors, autopct = '%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
# Subplots
X = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]
# Create subplots (2 rows, 1 column)
fig, (ax1, ax2) = plt.subplots(2, 1)

# Plot on the first subplot 
ax1.plot(X, y1, color='blue')
ax1.set_title('Plot 1: y=x^2')

# Plot on the second subplot 
ax2.plot(X, y2, color='green')
ax2.set_title('Plot 2: y=x^2 (inverted)')

# Display the plots
plt.tight_layout() # Automatically adjusts subplot parameters
plt.show()
```

```python
#boxplot
data = np.random.rand(100)
plt.boxplot(data, vert=False, patch_artist= True)
plt.title("boxplot")
plt.xlabel("values")
plt.show()
```

```python
import seaborn as sns
```

```python
#load the tips dataset
tips = sns.load_dataset('tips')
print(type(tips))
```

```python
tips.info()
```

```python
#ctreate a scatter plot using seaborn
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',palette='coolwarm')
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Filter numeric columns only
numeric_data = tips.select_dtypes(include=['number'])
correlation = numeric_data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Tips Dataset")
plt.show()
```
